```python
import csv
```

```python
streets={}
```

```python
with open('data.csv', 'r', encoding='cp1251') as f:
    line = f.readline()
    print(line)
```

```python
with open('data.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        print(row)
```

```python
%time
with open('data.csv', 'r', encoding="cp1251") as f:
    reader = csv.DictReader(f, delimiter=";")
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```
